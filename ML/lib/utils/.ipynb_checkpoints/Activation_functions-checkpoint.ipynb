```python
import numpy as np
import sys
# reference: https://en.wikipedia.org/wiki/Activation_function
class Sigmoid():
    def __init__(self):
        pass
    def function(self,z):
        return 1 / (1 + np.exp(-z))
    def gradient(self,z):
        return self.function(z) * (1-self.function(z))
    
class SoftMax():
    # i.e. Weighted sigmoid
    def __init__(self):
        pass
    def function(self,z):
        e_z = np.exp(z - np.max(z, axis=1, keepdims=True))
        return e_z / np.sum(e_z, axis=1, keepdims=True)
    def gradient(self,z):
        return self.function(z) * (1-self.function(z))    
    
class TanH():
    def __init__(self):
        pass
    def function(self,z):
        return 2 / (1 + np.exp(-2 * z)) - 1
    def gradient(self,z):
        return 1 - np.power(self.function(z), 2)

class ReLU():
    # z  |  z > 0
    # 0  |  z < 0
    def __init__(self): 
        pass 
    def function(self, z):
        return np.where(z >= 0, z, 0)

    def gradient(self, x):
        return np.where(z >= 0, z, 0)

class LeakyReLU():
    # z        |  z > 0
    # z*alpha  |  z < 0
    def __init__(self, alpha=0.2):
        self.alpha = alpha

    def function(self, z):
        return np.where(z >= 0, z, self.alpha * z)

    def gradient(self, x):
        return np.where(z >= 0, 1, self.alpha)    
```

```python
import matplotlib.pyplot as plt


sig   = Sigmoid()
wsig  = SoftMax()
tH    = TanH()
reLu  = ReLU()
lreLu = LeakyReLU()

vals = np.arange(-1,1,0.1)
# Visualize activation funcitons
plt.title('Activation Functions')
plt.plot(vals,sig.function(vals))
plt.plot(vals,tH.function(vals))
plt.plot(vals,reLu.function(vals))
plt.plot(vals,lreLu.function(vals))
plt.ylim([-1,1])
plt.legend(["Sigmoid","TanH","ReLu","LeakyReLu"])
plt.show()
```
