```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
from scipy.io import loadmat
%matplotlib inline
```

```python
## Principle Component Analysis
def pca(X):
    # normalize the features
    X = (X - X.mean()) / X.std()
    
    # compute the covariance matrix
    X = np.matrix(X)
    cov = (X.T * X) / X.shape[0]
    
    # perform SVD
    U, S, V = np.linalg.svd(cov)
    
    return U, S, V

def project_data(X, U, k):
    U_reduced = U[:,:k]
    return np.dot(X, U_reduced)

def recover_data(Z, U, k):
    U_reduced = U[:,:k]
    return np.dot(Z, U_reduced.T)
```

```python
## Using PCA for dimensionality reduction instead
faces = loadmat('data/ex7faces.mat')
X = faces['X']
face = np.reshape(X[3,:], (32, 32))  # looking at only 3rd image
plt.imshow(face)
plt.axis('off')
# Perform PCA to yield the top 100 Principle components
U, S, V = pca(X) 
```

```python
# PCA performs SVD on the data (using its covariance mat)
k = 1
PC1 = np.reshape(U[:,k], (32, 32))

# Eigenvectors
plt.imshow(PC1)
plt.title("Visualizing Principle Component: %d"%(k))
plt.axis('off')
plt.show()

# Eigenvalues
plt.title("Visualizing the Eigenvalue attributed to each PC")
lam = np.sqrt(S[:])
optPCs = np.argmax(-np.diff(lam) <= 0.012)
optPCs = np.argmax(-np.diff(lam) <= 0.002)
plt.plot([optPCs,optPCs],[0,np.amax(lam)],'k--')
plt.plot(lam)
plt.ylabel('Eigenvalue')
plt.xlabel('Principle Component #')
plt.show()
```

```python
print("The Qualitative method for feature reduction is to select all the PCs up until the elbow:", optPCs)
# Reprojecting image
Z = project_data(X, U, optPCs)
rec_face = recover_data(Z, U, optPCs)
print("Now, time to re-project the PC space (%d dim) to the orininal feature space (%d dim)"%(optPCs,X.shape[1]))

pcs_face = np.reshape(rec_face[3,:], (32, 32))  # looking at only 3rd image
plt.imshow(pcs_face)
plt.axis('off')
plt.show()

# Evaluate quantitatively (implement error function)
pcvals  = np.reshape(pcs_face,(1024))
orgvals = np.reshape(face,(1024))
err     = np.sqrt(np.sum(np.power((pcvals-orgvals),2)))
print("Reconstruction error:",err)
```
