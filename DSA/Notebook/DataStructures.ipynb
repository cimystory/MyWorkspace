```python
import ctypes ## Used for Dynamic Array

class DynamicArray(object):
    # similar to list
    
    def __init__(self):
        self.n = 0             # Count of elements in list
        self.capacity = 1      # Default Capacity
        self.A = self.make_array(self.capacity)
        
    def __len__(self):
        # returns length of array
        return self.n
    
    def __getitem__(self,k):
        # Retrieve element at index k
        # base case (k needs to lie within bounds of array)
        if not 0 <= k < self.n:
            return IndexError('K is out of bounds!')
        
        return self.A[k] 
        
    def append(self, ele):
        # base case (if capacity is reached, double size of array)
        if self.n == self.capacity:
            self._resize(2*self.capacity)
         #add element to end of list
        self.A[self.n] = ele
        self.n += 1
        
    def _resize(self,new_cap):
        #resize array (base condition of append)
        
        # make bigger array
        B = self.make_array(new_cap) 
        # repopulate new array with elements in old array
        for k in range(self.n): 
            B[k] = self.A[k]
   
        self.A = B 
        # Reset the capacity
        self.capacity = new_cap 
        
    def make_array(self,new_cap):
        # returns new array with new size
        return (new_cap * ctypes.py_object)()
```

```python
## Doubly Linked List
class Node(object):
    
    def __init__(self,value): 
        self.value     = value
        self.nextnode  = None 
        self.prevnode = None
    
class DoublyLinkedListNode(object):
    
    def __init__(self,node):
        self.tail      = None
        node.nextnode  = self.tail
        self.head      = node
        self.__length    = 1
                 
    def add(self,Node):
        self.__length    += 1
        tailNode = self.__findTail()
        tailNode.nextnode = Node
        Node.prevnode = tailNode
        return self
        
    def remove(self,Node):
        targetNode = self.__findNode(Node)
        if targetNode is None:
            return targetNode
        self.__length    -= 1
        tempnext = targetNode.nextnode
        tempprev = targetNode.prevnode
        tempprev.nextnode  = tempnext
        tempnext.prevnode  = tempprev
        return targetNode
    
    def __findTail(self):
        curnode = self.head
        while curnode.nextnode != None:
            curnode = curnode.nextnode
        return curnode
    
    def __findNode(self,targetNode):
        curnode = self.head
        while curnode.nextnode != None and curnode != targetNode:
            curnode = curnode.nextnode
        return curnode
    
    def printNodesForward(self):
        curnode = self.head
        print("Node: ",curnode,"Value: ",curnode.value)
        while curnode.nextnode != None:
            curnode = curnode.nextnode
            print("Node: ",curnode,"Value: ",curnode.value)
    
    def printNodesBackward(self):
        curnode = self.__findTail()
        print("Node: ",curnode,"Value: ",curnode.value)
        while curnode.prevnode != None:
            curnode = curnode.prevnode
            print("Node: ",curnode,"Value: ",curnode.value)
            
    def get_length(self):
        return self.__length
```

```python
class Stack(object):
    # Fill out the Stack Methods here
    def __init__(self):
        self.contents = []
        
    def isempty(self):
        return self.contents == []
    
    def push(self,value):
        self.contents.append(value)
        
    def pop(self):
        if self.isempty():
            return None
        content       = self.contents[-1]
        self.contents = self.contents[:-1]
        return content
    
    def peek(self):
        if self.isempty():
            return None
        return self.contents[-1]
    
    def size(self):
        return len(self.contents)
    
    def printcontents(self):
        for val in self.contents:
            print(val)
```

```python
class Queue(object):
    def __init__(self):
        self.contents = []       
        
    def isempty(self):
        return self.contents == []
    
    def Enqueue(self,value):
        self.contents.append(value)

    def Dequeue(self):
        if self.isempty():
            return None
        content       = self.contents[0]
        self.contents = self.contents[1:]
        return content
    
    def size(self):
        return len(self.contents)
            
```

```python
class Deque(object):
    
    def __init__(self):
        self.contents = []

    def isempty(self):
        return self.contents == []

    def addFront(self,value):
        self.contents.insert(0,value)
        
    def addBack(self,value):
        self.contents.append(value)
        
    def removeFront(self):
        return self.contents.pop(0)
    
    def removeBack(self):
        return self.contents.pop()

    def size(self):
        return len(self.contents)
```
