```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as statistics
from matplotlib.patches import Ellipse
import math
```

```python
# Helper function
## NOTE: I use scipy package instead
def gaussian_3dplot(rv,xrange,yrange):
    x, y = np.mgrid[xrange[0]:xrange[1]:.01, yrange[0]:yrange[1]:.01]
    pos = np.empty(x.shape + (2,))
    pos[:, :, 0] = x
    pos[:, :, 1] = y
    plt.contourf(x, y, rv.pdf(pos))
    plt.xlabel("X-Dimension")
    plt.ylabel("Y-Dimension")
    plt.title("Multi-Variate Covariance Matrix")
    plt.show()
    return 
def plot_cov_contour(x,y,cov_mat,color = 'k'):
    U,s,v = np.linalg.svd(cov_mat)
    orientation = math.atan2(U[1,0],U[0,0])
    a = np.degrees(orientation)
    w = math.sqrt(s[0]) * 2.
    h = math.sqrt(s[1]) * 2.
    ell = Ellipse(xy=(np.mean(x), np.mean(y)),
                  width=w, height=h, angle=a,
                  edgecolor=color, fc='None', lw=2)
    ell.set_facecolor('none')
    return ell
```

```python
# Vizualizing Distribution and their properties
# Random Distribution
print("\tVisualizing a Random distribution")
    # Generating Data
n = 1000
rdist = 10*np.random.rand(n)
hist,edges = np.histogram(rdist,bins='auto')
    # Stats
stats = (np.mean(rdist),np.std(rdist))
mcnt  = [np.amax(hist)+10,np.amax(hist)+10]
sigs = ((stats[0]-stats[1],stats[0]+stats[1]),(stats[0]-2*stats[1],stats[0]+2*stats[1]))
# Visualize
plt.bar(edges[:-1],hist,align='edge',edgecolor='k',)
plt.bar(stats[0],mcnt[0],align='edge',width=0.3,color='k')
plt.bar(sigs[0],mcnt,align='edge',width=0.2,color='r')
plt.bar(sigs[1],mcnt,align='edge',width=0.1,color='r')
plt.ylim([0,np.amax(hist)])
plt.title('Random Distribution')
plt.ylabel('Count')
plt.xlabel('Value')
plt.show()
print("mean: %.2f, std: %.2f"%stats)
```

```python
# Normal Distribution
print("\tVisualizing a Normal distribution")
    # Generating Data
ndist = np.random.normal(loc=stats[0], scale=stats[1], size=n) # Creating normal distribution
    # STATS
nhist,nedges = np.histogram(ndist,bins='auto')
nstats = (np.mean(ndist),np.std(ndist))
nsigs = ((nstats[0]-nstats[1],nstats[0]+nstats[1]),(nstats[0]-2*nstats[1],nstats[0]+2*nstats[1]))
    # PDF (2 methods, norm & kernel)
x    = np.linspace(nedges[0],nedges[-1],200)
npdf = statistics.norm(nstats[0],nstats[1]).pdf(x) # Generating normal PDF, Method1
kde  = statistics.gaussian_kde(ndist)              # Generating normal PDF, Method2
mcnt  = [np.amax(npdf*n)+10,np.amax(npdf*n)+10]

# Visualize
plt.bar(nedges[:-1],nhist,align='edge',edgecolor='k')
plt.plot(x,npdf*n,'c--')
plt.plot(x,kde(x)*n,'c-')
plt.bar(nstats[0],mcnt[0],align='edge',width=0.3,color='k')
plt.bar(nsigs[0],mcnt,align='edge',width=0.2,color='r')
plt.bar(nsigs[1],mcnt,align='edge',width=0.1,color='r')
plt.ylim([0,np.amax(mcnt)])
plt.title('Normal Distribution')
plt.ylabel('Count')
plt.xlabel('Value')
plt.show()
print("mean: %.2f, std: %.2f"%nstats)
    # CDF
amnt1  = statistics.norm(stats[0],nstats[1]).cdf(nsigs[0])   # Using normal CDF, To evaluate stats in range
amnt2  = statistics.norm(stats[0],nstats[1]).cdf(nsigs[1])   # Using normal CDF, To evaluate stats in range
print("\t How much of the data lies between sigma values")
print("Percent of data that lies within 1 sig: %.2f"%np.diff(amnt1))
print("Percent of data that lies within 2 sig: %.2f"%np.diff(amnt2))
```

```python
# Exponential Distributions
# Generating Data
ndist = np.random.exponential(size=n, scale=1/stats[0]) # Creating normal distribution
    # STATS
nhist,nedges = np.histogram(ndist,bins=200)
nstats = (np.mean(ndist),np.std(ndist))
nsigs = ((nstats[0]-nstats[1],nstats[0]+nstats[1]),(nstats[0]-2*nstats[1],nstats[0]+2*nstats[1]))
    # PDF (2 methods, norm & kernel)
x       = np.linspace(nedges[0],nedges[-1],200)
expfit  = statistics.expon(scale = 1/stats[0]).pdf(x)
kde     = statistics.gaussian_kde(ndist)              # Generating normal PDF, Method2
mcnt    = [np.amax(nhist)+10,np.amax(nhist)+10]
# Visualize
plt.hist(ndist,edgecolor='k')
plt.plot(x,expfit*100,'c-')
plt.bar(nstats[0],mcnt[0],align='edge',width=0.05,color='k')
#plt.ylim([0,np.amax(mcnt)])
plt.title('Exponential Distribution')
plt.ylabel('Count')
plt.xlabel('Value')
plt.show()
print("Lambda: ",1/stats[0])
```

```python
# Exponential Distributions
# Generating Data
xrange = (2-10,2+10)
yrange = (17-10,17+10)
mu = [2.0, 17.0]
P = [[10., 0.], 
     [0., 4.]]
nrv = statistics.multivariate_normal(mu, P) # Creating normal distribution
gaussian_3dplot(nrv,xrange,yrange)
```

```python
# Exploring Multi-Variate Correlations
height = [60, 62, 63, 65, 65.1, 68, 69, 70, 72, 74]
weight = [95, 120, 127, 119, 151, 143, 173, 171, 180, 210]
P0 = np.cov(height,weight)

    
ax = plt.subplot(111)
plt.scatter(height, weight)
ell1 = plot_cov_contour(height,weight,P0)
ax.add_patch(ell1)

plt.title('Track Team Height vs. Weight')
plt.xlabel('Height (in)')
plt.ylabel('Weight (lbs)')
plt.show()
```

```python
# Visualizing improvement of Variance across three separate data sets
posx = 10
posy = 10
mu = (posx,posy)
P0 = [[6, 0], [0, 6]]
P1 = [[2, 1.9], [1.9, 2]]
P3 = [[2, -1.9], [-1.9, 2.2]]
# Generate Synthetic Data 
x1,y1 = np.random.multivariate_normal(mu, P0, 20).T # Creating normal distribution
x2,y2 = np.random.multivariate_normal(mu, P1, 20).T # Creating normal distribution
x3,y3 = np.random.multivariate_normal(mu, P3, 20).T # Creating normal distribution

ax = plt.subplot(111)
plt.scatter(x1, y1,color='b')
plt.scatter(x2, y2,color='g')
plt.scatter(x3, y3,color='r')

ell1 = plot_cov_contour(x1,y1,P0,'b')
ell2 = plot_cov_contour(x2,y2,P1,'g')
ell3 = plot_cov_contour(x3,y3,P3,'r')

ax.add_patch(ell1)
ax.add_patch(ell2)
ax.add_patch(ell3)

plt.title('Track Aircraft position:\nSensor Readings:  x vs. y')
plt.legend(['Sensor 1','Sensor 2','Sensor 3'])
plt.xlabel('X Position (m)')
plt.ylabel('Y Position (m)')
plt.show()
print("The intersection, formed by multiplying the prior and measurement, is a new Gaussian. \nThe shapes reflects the geometry of the problem. This allows us to triangulate on the aircraft, \n\tresulting in a very accurate estimate")
```
