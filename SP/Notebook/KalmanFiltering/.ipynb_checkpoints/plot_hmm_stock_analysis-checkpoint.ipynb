
Gaussian HMM of stock data
--------------------------

This script shows how to use Gaussian HMM on stock price data from
Yahoo! finance. For more information on how to visualize stock prices
with matplotlib, please refer to ``date_demo1.py`` of matplotlib.



```python
from __future__ import print_function

import datetime

import numpy as np
import matplotlib.pyplot as plt
from hmmlearn.hmm import GaussianHMM
from urllib.request import urlopen
from bs4 import BeautifulSoup as bs
from datetime import datetime
print(__doc__)
```

```python
def get_historical_data(name, number_of_days):
    data = data = {'Date':[],'Open':[],'Close':[],'Volume':[]}
    url = "https://finance.yahoo.com/quote/" + name + "/history/"
    print(url)
    rows = bs(urlopen(url).read(),"html5lib").findAll('table')[1].tbody.findAll('tr')

    for each_row in rows:
        divs = each_row.findAll('td')
        if divs[1].span.text  != 'Dividend': #Ignore this row in the table
            #I'm only interested in 'Open' price; 
            # ... For other values, play with divs[1 - 5]
            data['Date'].append(datetime.strptime(divs[0].span.text,'%b %d, %Y'))
            data['Open'].append(float(divs[1].span.text.replace(',','')))
            data['Close'].append(float(divs[4].span.text.replace(',','')))
            data['Volume'].append(float(divs[6].span.text.replace(',','')))

    return data
```

Get quotes from Yahoo! finance



```python
# Return the financial data for 
google = get_historical_data("INTC", 730)
plt.plot(google['Date'],google['Open'])
plt.plot(google['Date'],google['Close'])
plt.legend(['Open','Close'])
#plt.plot(google['Date'],google['Volume'])
plt.show()


# Take diff of close value.
dClose  = np.diff(google['Close'])
cvolume = np.array(google['Volume'][1:])
cdates  = np.array(google['Date'][1:])

# Pack diff and volume for training.
X = np.column_stack([dClose, cvolume])
```

Run Gaussian HMM



```python
print("fitting to HMM and decoding ... ", end="")

# Make an HMM instance and execute fit
model = GaussianHMM(n_components=4, covariance_type="diag", n_iter=1000).fit(X)

# Predict the optimal sequence of internal hidden state
hidden_states = model.predict(X)

print("done")
```

Print trained parameters and plot
