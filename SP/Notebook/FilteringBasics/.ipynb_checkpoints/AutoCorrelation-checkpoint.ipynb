```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
import sounddevice as sd
import scipy.io.wavfile
import scipy
from scipy import signal
import math
import pandas as pd
from datetime import datetime
```

```python
# Helper Function for computing correlations
def serialCorr(wave1,wave2,lag=1):
    n = len(wave1)
    y1 = wave1[lag:]
    y2 = wave2[:n-lag]
    pcorr_mat = np.corrcoef(y1,y2)
    return pcorr_mat[0,1]

def autoCorr(wave1,wave2):
    n = len(wave1)
    autocorr_mat = np.zeros([n//2,1])
    rlag = range(0,n//2)
    for templag in rlag:
        autocorr_mat[templag,0] = serialCorr(wave1,wave2,lag=templag)
    return autocorr_mat[:,0]

# Helper function to generating a Waveforms
# Sine Wave
def genSineWave(amp = 1,freq = 1,off = 0,dur = 1,frate = 10000):
    dt        = 1/frate
    ts = np.arange(0,dur+dt,dt)
    t  = np.arange(0,len(ts))
    return amp*np.sin((2*np.pi*freq*t/frate)+off),ts

# Noise
def GenerateBrownianNoise(amp=1,dur=1,frate=10000):
    ts   = np.arange(0,dur+1/frate,1/frate)
    dsig = np.random.uniform(-1, 1, len(ts))
    sig  = np.cumsum(dsig)
    noise  = sig - np.mean(sig)
    noise /= np.amax(noise)
    noise *= amp
    return noise,ts

def GenerateUncorrelatedNoise(amp=1,dur=1,frate=10000):
    ts = np.arange(0,dur+1/frate,1/frate)
    result = np.random.uniform(-amp, amp, len(ts))
    return result,ts

def GenerateBetaNoise(signal,fq,Beta = 1):
    f = fq ** (Beta/2)
    f[0] = 1
    return signal/f

# Helper Function for converting waves to fft power spectrums back and forth
def make_spectrum(datain, ratein):
    n = len(datain)
    # power spectral density
    hsout = np.fft.rfft(datain)
    # frequency spectrum
    freq = np.fft.rfftfreq(n, d=1./ratein)
    return hsout, freq

def make_wave(hsin):
    wave = np.fft.irfft(hsin)
    return wave
```

```python
## PEARSON CORRELATION
# Generate Sine waves with different offset
off1 = 0
off2 = 1#math.radians(1)
wave1,ts1 = genSineWave(amp = 1,freq = 440,off = off1, dur = 0.01,frate = 10000)
wave2,ts2 = genSineWave(amp = 1,freq = 440,off = off2, dur = 0.01,frate = 10000)
plt.plot(ts1,wave1)
plt.plot(ts2,wave2)
plt.show()
# Pearson Correlation computed as -->  sum[(xi-ux)(yi-uy)]/(N[sigx][sigy])
pcorr_mat = np.corrcoef(wave1,wave2)
print("Pearson Correlation Matrix: \n",pcorr_mat) 
```

```python
## SERIAL CORRELATION
bnoise,bts   = GenerateBrownianNoise()
ucnoise,ucts = GenerateUncorrelatedNoise()
scorr_bnoise  = serialCorr(bnoise,bnoise,lag=1)
scorr_ucnoise = serialCorr(ucnoise,ucnoise,lag=1)
print("\tAs is expected: There is higher auto-correlation in Brown vs Uncorrelated Noise")
print("Brownian Noise Serial Correlation with lag=1: ",scorr_bnoise)
print("Uncorrelated Noise Serial Correlation with lag=1: ",scorr_ucnoise)
# Compare Serial correlation [lag = 1] of noise generated with varying Beta values [0-2]
beta  = np.arange(0,2.2,0.2)
scorr_array = np.zeros([len(beta),1])
frate = 10000
for bidx in range(len(beta)):
    bval = beta[bidx]
    tempspec,fq = make_spectrum(ucnoise.copy(), frate)
    new_spec    = GenerateBetaNoise(tempspec,fq,Beta = bval)
    tempwave    = make_wave(new_spec)
    scorr_array[bidx] = serialCorr(tempwave,tempwave,lag=1)
plt.plot(beta,scorr_array)
plt.title('Serial Correlation (lag = 1) vs Noise Beta Value')
plt.ylabel('Correlation')
plt.xlabel('Beta')
plt.show()
```

```python
## AUTO CORRELATION
ucnoise,ucts = GenerateUncorrelatedNoise()
# Compare Serial correlation [lag = 1] of noise generated with varying Beta values [0-2]
beta  = np.arange(0,2.5,0.5)
xcorr_array = np.zeros([len(ucnoise)//2,len(beta)])
frate = 10000
for bidx in range(len(beta)):
    bval = beta[bidx]
    tempspec,fq = make_spectrum(ucnoise.copy(), frate)
    new_spec    = GenerateBetaNoise(tempspec,fq,Beta = bval)
    tempwave    = make_wave(new_spec)
    tempxcorr   = autoCorr(tempwave,tempwave)
    xcorr_array[:,bidx] = tempxcorr
plt.plot(xcorr_array)
plt.title('Auto Correlation for different Noise Models')
plt.ylabel('Correlation Coefficient')
plt.xlabel('Lag')
plt.xlim([0,1000])
plt.legend(beta)
plt.show()
print("There is a stronger presence of long-range dependence for noise models that are constructed from a higher Beta Value")
print("This indicates that current values of the signal are more dependent on preceding values")
```

```python
## ANALYZING AN AUDIO SIGNAL
# Load audio signal
datafolder = '/Users/maxmaguire/Desktop/PostGradWork/DataFolder/SampleAudioFiles/'
fname = datafolder + '28042__bcjordan__voicedownbew.wav'
rate, data = scipy.io.wavfile.read(fname)
ts = np.full([1,len(data)-1],1/rate)
ts = np.cumsum(ts)
ts = np.insert(ts,0,0)
#sd.play(data,rate)
# Visualize at STFT and Power Spectrum across whole audio file (bird chirping)
nsegLength = 512
f, t, Zxx = signal.stft(data, rate, nperseg=nsegLength)
ldata   = np.all([ts >= 0.2,ts<= 0.21],axis=0)
sampts   = ts[ldata]
sampdata = data[ldata]
spec,fq = make_spectrum(sampdata, rate)

plt.subplot(121)
plt.pcolormesh(t, f, np.abs(Zxx),cmap='jet')
plt.ylim([0,4000])
plt.title('Spectrogram')
plt.ylabel('Frequency')
plt.xlabel('Time (s)')
plt.subplot(122)
plt.plot(fq,np.power(np.abs(spec),2))
plt.xlabel('Frequency (hz)')
plt.ylabel('Amplitude')
plt.title('Power Spectrum\n to = 0.2,dur = 0.01')
plt.xlim([0,900])
plt.show()
peakfq = fq[np.argmax(np.power(np.abs(spec[1:]),2))+1]
print("The length of the segment is",len(sampdata)," samples at a frame rate of ",rate," Hz, so the frequency resolution is 100 Hz")
print("By looking at the power density spectrum, we see peak frequency recorded at ",peakfq, "(hz)")
# Compare that with autocorrelation Method
```

```python
xcorr = autoCorr(sampdata,sampdata)
plt.plot(xcorr)
plt.xlabel('Lag (sample)')
plt.ylabel('xCorr')
plt.title('Signal\n to = 0.2,dur = 0.01')
plt.show()
maxidx = np.argmax(xcorr[1:])+1
print("Auto correlation peaks at a frequency of ",maxidx," samples")
print("That is equivalent to a time lag of ", maxidx/rate,"(s)")
print("...Which is ",rate/maxidx, "(hz)")
print("################################################################")
scorrData1 = serialCorr(sampdata,sampdata,lag=101)
scorrData2 = serialCorr(sampdata,sampdata,lag=110)
print("Now compare the peak freq amplitude obtained looking at the power density spectrum vs autocorrelation: ")
print("Power density spectrum: 437 (hz) {~101 samples}",", Correlation: ",scorrData1)
print("Power density spectrum: 400 (hz) {~110 samples}",", Correlation: ",scorrData2)
```
