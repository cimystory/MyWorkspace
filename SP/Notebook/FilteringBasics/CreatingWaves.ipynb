```python
import time
import numpy as np
import matplotlib.pyplot as plt
import sounddevice as sd
```

```python
# Function to create sinewave from timepoints/freq
def createSinWave(amp,t,freq,fs,offset = 0):
    return amp*np.sin(2*np.pi*freq*t/fs+offset)
# Function to create cosinewave from timepoints/freq
def createCosWave(amp,t,freq,fs,offset = 0):
    return amp*np.cos(2*np.pi*freq*t/fs+offset)
```

```python
# Signal specs
amp1   = 1
amp2   = 2
freq1 = 440   #freq hz
freq2 = 220  #freq hz
fs    = freq1*10 #sampling rate hz
off1  = 0
off2  = np.pi/4
t1    = np.arange(freq1*4) # samples
ts    = t1/fs              # time
```

```python
# Generate waveforms
wav1 = createSinWave(amp1,t1,freq1,fs)
wav2 = createCosWave(amp2,t1,freq2,fs)
```

```python
# Visualize waveforms individually
plt.figure(1)
plt.subplot(211)
plt.title("Simple Sine waveforms")
plt.plot(ts,wav1,'ro-')
plt.ylabel("Amplitude")
plt.subplot(212)
plt.plot(ts,wav2,'bo-')
plt.ylabel("Amplitude")
plt.xlabel("Time (s)")
plt.show()
```

```python
# Visualize waveforms mixed together
wavtot = wav1 + wav2
# Listen to sound
sd.play(wavtot,fs)
plt.figure(2)
plt.plot(ts,wavtot)
plt.title("Mix of Simple Sine waveforms")
plt.ylabel("Amplitude")
plt.xlabel("Time (s)")
plt.show()
```
